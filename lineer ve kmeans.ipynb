```python
def a_ve_b_yi_bul(x, y):
    a = sum((x - x.mean()) * (y - y.mean())) / sum((x - x.mean())** 2)
    b = y.mean() - a * x.mean()
    return a, b

a, b = a_ve_b_yi_bul(ozgenin_verisi["life_exp"], ozgenin_verisi["gdp_usd"])
print(a, b)
```

```python
df = pd.read_csv(dosya_yolu)
x = df[["life_exp", "gdp_usd"]]
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters= 3)
kmeans.fit(x)
centers = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
plt.scatter(x.iloc[:, 0], x.iloc[:, 1], c=labels, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75)
plt.xlabel("Life Expectancy")
plt.ylabel("GDP (USD)")
plt.title("Yıllara Göre Kişi Başına Gelir ve Yaşam Beklentisi")
plt.show()
```
